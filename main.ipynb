# Tratamento dos Dados


---



```python
dados.isnull().sum()
```

```python
df = dados.fillna(0)
```

```python
df.isnull().sum()
```

```python
df.query('Artist == 0')
```

```python
df.query('Artist == 0').index
```

```python
registros_a_remover = df.query('Artist == 0').index
df.drop(registros_a_remover, axis=0, inplace=True)
```

```python
df.query('Artist == 0')
```

```python
df.drop('TIDAL Popularity', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Ajustando o Data Type das colunas
df.info()
```

```python
import numpy as np
```

```python
df['Track Score'] = df['Track Score'].astype(np.float64)
```

```python
df['Spotify Popularity']
```

```python
df.columns
```

```python
df['All Time Rank'] = df['All Time Rank'].apply(lambda x : x.replace(',', '').strip())
```

```python
df['All Time Rank'] = df['All Time Rank'].astype(np.int64)
```

```python
df['Spotify Streams'] = df['Spotify Streams'].astype(str)
```

```python
df['Spotify Streams'] = df['Spotify Streams'].apply(lambda x : x.replace(',', '').strip())
```

```python
df['Spotify Streams'] = df['Spotify Streams'].astype(np.int64)
```

```python
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype(str)
```

```python
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].apply(lambda x : x.replace(',', '').strip())
```

```python
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype(np.int64)
```
